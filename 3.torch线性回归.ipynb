```python
from  torch import optim,nn 
```

```python
class Lr(nn.Module):
    def __init__(self):
        super(Lr,self).__init__()
        self.linear = nn.Linear(1,1)
    def forward(self, x):
        out = self.linear(x)
        return out
```

```python
modle = Lr()
loss = nn.MSELoss()
optimizer = optim.SGD(modle.parameters(), lr = 1e-3)
```
