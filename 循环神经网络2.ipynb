```python
import torch
```

```python
input_size = 4
hidden_size = 4
batch_size = 1
num_layers=1
num_embeddings = 10
num_class = 4
```

```python
idx2char = ['e','h','l','o']
x_data = [[1,0,2,2,3]]
y_data = [3,1,2,3,2]
inputs = torch.LongTensor(x_data)
labels = torch.LongTensor(y_data)
```

```python
inputs
```

```python
class Moudle(torch.nn.Module):
    def __init__(self,input_size,hidden_size,batch_size,num_layers,num_embeddings):
        super().__init__()
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.batch_size = batch_size
        self.num_layers = num_layers
        self.num_embeddings = num_embeddings
        self.emb = torch.nn.Embedding(input_size,num_embeddings)
        self.rnn = torch.nn.RNN(input_size=self.num_embeddings,hidden_size=self.hidden_size,num_layers=self.num_layers,batch_first=True)
        self.Linear = torch.nn.Linear(hidden_size,num_class)
    def forward(self,input):
        input = self.emb(input)
        print('input',input)
        h0 = torch.zeros(self.num_layers,self.batch_size,self.hidden_size)
        output,h0 = self.rnn(input,h0)
        output = self.Linear(output) 
        return output.view(-1,num_class)
```

```python
modle = Moudle(input_size,hidden_size,batch_size,num_layers,num_embeddings)
creterion = torch.nn.CrossEntropyLoss()
optim = torch.optim.Adam(modle.parameters(),lr=0.05)
```

```python
for i in range(15):
    optim.zero_grad()
    output = modle(inputs)
    print('output',output)
    print('labels',labels)
    loss = creterion(output,labels)
    print('loss', loss.item())
    loss.backward()
    optim.step()
```
